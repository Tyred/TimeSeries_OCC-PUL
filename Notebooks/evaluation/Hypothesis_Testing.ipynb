# Imports

```python
import Orange
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Read csv file with result metrics for each algorithm for all binary datasets in UCR Time Series Archive

```python
dicio = pd.read_csv('f1.csv')
```

```python
for col in dicio.columns.to_list():
    dicio[col] = dicio[col].str.rstrip('%').astype('float') / 100.0
```

```python
dicio
```

# Hypothesis Testing



```python
avranks = dicio.rank(axis=1,ascending=True).mean().values
cd = Orange.evaluation.compute_CD(avranks, dicio.shape[0], 
                                  alpha="0.05", test="nemenyi")
```

```python
Orange.evaluation.graph_ranks(avranks, dicio.keys(), 
                              cd=cd, width=7, textspace=1.5)
plt.savefig('nemenyi.pdf', bbox_inches='tight')
plt.show()
```

```python
avranks = dicio.rank(axis=1,ascending=False).mean().values
cd = Orange.evaluation.compute_CD(avranks, dicio.shape[0], 
                                  alpha="0.05", test="nemenyi")

Orange.evaluation.graph_ranks(avranks, dicio.keys(), 
                              cd=cd, width=7, textspace=1.5)
plt.savefig('nemenyi.pdf', bbox_inches='tight')
plt.show();
```
